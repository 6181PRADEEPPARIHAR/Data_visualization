```python
#  line chart for Date of Joining
df['DateOfJoining'] = pd.to_datetime(df['DateOfJoining'])
joinings_by_date = df.groupby(df['DateOfJoining'].dt.year)['EmployeeCode'].count()
joinings_by_date.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Joinings')
plt.title('Trend of Employee Joinings Over Time')
```

## **Resultant Observation from above**

*   The line chart titled "Trend of Employee Joinings Over Time" displays the number of employee joinings each year.

```python
# Bivariate analysis example: Gender vs. MaritalStatus
cross_tab = pd.crosstab(df["Gender"], df["MaritalStatus"])
cross_tab.plot(kind="bar", stacked=True)
plt.title("Bivariate Analysis: Gender vs. MaritalStatus")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Bivariate analysis example: DateOfBirth vs. Salary
sns.scatterplot(x="DateOfBirth", y="Salary", data=df)
plt.title("Bivariate Analysis: DateOfBirth vs. Salary")
plt.xlabel("DateOfBirth")
plt.ylabel("Salary")
plt.show()

# Bivariate analysis example: Department vs. Skill
cross_tab = pd.crosstab(df["Department"], df["Skill"])
cross_tab.plot(kind="bar", stacked=True)
plt.title("Bivariate Analysis: Department vs. Skill")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()

# Bivariate analysis example: EmployeeStatus vs. ResignReason
cross_tab = pd.crosstab(df["EmployeeStatus"], df["ResignReason"])
cross_tab.plot(kind="bar", stacked=True)
plt.title("Bivariate Analysis: EmployeeStatus vs. ResignReason")
plt.xlabel("EmployeeStatus")
plt.ylabel("Count")
plt.show()
```

## **Gender vs. MaritalStatus:**
This analysis uses a bar chart (stacked) to show the distribution of marital status among different genders.
It helps visualize how marital status is distributed across genders in your dataset.

## **DateOfBirth vs. Salary:**
This analysis uses a scatter plot to show the relationship between an employee's date of birth and their salary.
It allows you to explore whether there's any correlation or pattern between age and salary.

## **Department vs. Skill:**
This analysis uses a stacked bar chart to display the distribution of skills across different departments.
It helps you understand which skills are predominant in each department or if there's any overlap between skills in different departments.

## **EmployeeStatus vs. ResignReason:**


This analysis uses a stacked bar chart to visualize the reasons for employee resignations across different employee statuses.
